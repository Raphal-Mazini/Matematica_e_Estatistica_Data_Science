### Exercício 1 - Representação Geométrica de Um Vetor

Neste exercício você deve representar geometricamente o vetor:

v = np.array([-1,2])

```python
# Cria um vetor
v = np.array([-1,2])

# Plot do vetor
plt.arrow(0, 0, v[0], v[1], head_width = .5, width = .1)
plt.plot(0, 0, 'ko', markerfacecolor = 'k', markersize = 7)

# Adiciona as linhas dos eixos
plt.plot([-3,3], [0,0], '--', color = [.8,.8,.8], zorder = -1)
plt.plot([0,0], [-3,3], '--', color = [.8,.8,.8], zorder = -1)

# Formnata o Plot
plt.axis('square')
plt.axis([-3,3,-3,3])
plt.xlabel('$v_0$')
plt.ylabel('$v_1$')
plt.title('Plot do Vetor')
plt.show()
```

### Exercício 2 - Representação Geométrica da Soma dos Vetores

Neste exercício você deve representar geometricamente os 2 vetores abaixo e o vetor resultante da soma entre eles.

v = np.array([1,2])

w = np.array([4,-6])

```python
# Define os vetores
v = np.array([1,2])
w = np.array([4,-6])

# Soma dos vetores
somaVetores = v + w

# Tamanho da figura
plt.figure(figsize = (6,6))

# Plot dos 3 vetores
a1 = plt.arrow(0, 
               0, 
               v[0], 
               v[1], 
               head_width = .3, 
               width = .1, 
               color = 'r', 
               length_includes_head = True)

a2 = plt.arrow(v[0], 
               v[1], 
               w[0], 
               w[1], 
               head_width = .3, 
               width = .1, 
               color = 'b', 
               length_includes_head = True)

a3 = plt.arrow(0, 
               0, 
               somaVetores[0], 
               somaVetores[1], 
               head_width = .3, 
               width = .1, 
               color = 'g', 
               length_includes_head = True)

# Formata o plot
plt.grid(linestyle = '--', linewidth = .5)
plt.axis('square')
plt.axis([-6, 6, -6, 6])
plt.legend([a1, a2, a3],['v', 'w', 'v+w'])
plt.title('Vetores $\mathbf{v}$, $\mathbf{w}$ e $\mathbf{v+w}$')
plt.show()
```

### Exercício 3 - Representação Geométrica da Subtração de Vetores

Neste exercício você deve representar geometricamente os 2 vetores abaixo e o vetor resultante da subtração entre eles.

v = np.array([1,2])

w = np.array([4,-6])

```python
# Define os vetores
v = np.array([1,2])
w = np.array([4,-6])

# Subtração dos vetores
difVetores = v - w

# Tamanho da figura
plt.figure(figsize=(6,6))

# Plot dos 3 vetores
a1 = plt.arrow(0, 
               0, 
               v[0], 
               v[1], 
               head_width = .3, 
               width = .1, 
               color = 'm', 
               length_includes_head = True)

a2 = plt.arrow(0, 
               0, 
               w[0], 
               w[1], 
               head_width = .3, 
               width = .1, 
               color = 'y', 
               length_includes_head = True)

a3 = plt.arrow(w[0], 
               w[1], 
               difVetores[0], 
               difVetores[1], 
               head_width = .3, 
               width = .1, 
               color = 'black', 
               length_includes_head = True)

# Formata o plot
plt.grid(linestyle = '--', linewidth = .5)
plt.axis('square')
plt.axis([-6, 6, -6, 6])
plt.legend([a1, a2, a3],['v', 'w', 'v-w'])
plt.title('Vetores $\mathbf{v}$, $\mathbf{w}$ e $\mathbf{v-w}$')
plt.show()
```
